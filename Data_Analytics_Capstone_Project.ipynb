```python
df_JobRole = pd.DataFrame(columns=["Job Role", "% of EE's leaving"])
i=0
for field in list(df['JobRole'].unique()):
    ratio = df[(df['JobRole']==field)&(df['Attrition']=="Yes")].shape[0] / df[df['JobRole']==field].shape[0]
    df_JobRole.loc[i] = (field, ratio*100)
    i += 1
 
```

```python
#print("In {}, the ratio of leavers is {:.2f}%".format(field, ratio*100))    
df_JR = df_JobRole.groupby(by="Job Role").sum()
df_JR
```

```python
# Yes/No Attrition
atrrition_yes_df = df[df['Attrition']=='Yes']
atrrition_no_df = df[df['Attrition']=='No']
```

```python
# Count of Attrition Yes and No's. 
job_role_yes = atrrition_yes_df['JobRole'].value_counts()
job_role_no = atrrition_no_df['JobRole'].value_counts()

fig, ax1 = plt.subplots(figsize=(20, 6))
# Pie chart with wide wedgeprop for JobRole
ax1.pie((job_role_yes*100/job_role_no), wedgeprops=dict(width=0.6), labels=(job_role_yes/job_role_no).index,autopct='%1.1f%%', startangle=180);
ax1.text(0., 0., '', horizontalalignment='center', verticalalignment='center')
ax1.axis('equal')

# Display charts
plt.show()
```

```python
# Create combo chart
fig, ax2 = plt.subplots(figsize=(19.5,15))
color = 'tab:blue'

# Create Bar Plot to show Monthly income and Job role 
ax2.set_xlabel('JobRole', fontsize=20)
ax2.set_ylabel('MonthlyIncome', fontsize=20, color=color)
ax2 = sns.barplot(x='JobRole', y='MonthlyIncome', data = df, palette='ch:s=.30,rot=-.30')
ax2.tick_params(axis='y')

# Calculate and add Attrition rates
attrition_rate = pd.DataFrame(pd.crosstab(df['JobRole'],df['Attrition']))
attrition_rate.insert(2, 'Attr%', value = (job_role_yes/(job_role_no + job_role_yes) *100))

# Specify that we want to share the same x-axis
ax3 = ax2.twinx()
color = 'tab:orange'

# Line plot to display the Attrition % by Job Role 
ax3.set_ylabel('% Attrition', fontsize=20, color=color)
ax3 = sns.lineplot(x='JobRole', y='Attr%', data= attrition_rate, sort=False, color=color)
ax3.tick_params(axis='y', color=color)

plt.show()
```

```python
##Attrition is highest within EE's without Stock Options or at an option level of 0 when compared to higher levels of options.
df.groupby(['StockOptionLevel'])['Attrition'].count().plot(kind = 'barh')
```

```python
# StockOptionLevel value counts correlated to Y/N Attrition
stock_option_count = df['StockOptionLevel'].value_counts()
stockoption_yes = atrrition_yes_df['StockOptionLevel'].value_counts()
stockoption_no = atrrition_no_df['StockOptionLevel'].value_counts()

# Calculate and add Attrition rates
stock_count = pd.DataFrame(pd.crosstab(df['StockOptionLevel'],df['Attrition']))
stock_count.insert(2, 'Attrition%', value = (stockoption_yes/stockoption_no *100))
stock_count.insert(2, 'Count', value = stock_option_count)

stock_count
```

```python
# Create DataFrame for valid interpretation of 'StockOptionLevel' input
stock_att=df.groupby(['StockOptionLevel','Attrition']).apply(lambda x:x['StockOptionLevel'].count()).reset_index(name='Count')

stock_att
```

```python
# Create combo chart
fig, ax1 = plt.subplots(figsize=(10,5))
color = 'tab:blue'

# Bar plot creation
ax1.set_xlabel('StockOptionLevel', fontsize=20)
ax1.set_ylabel('Count', fontsize=15, color=color)
ax1 = sns.barplot(x='StockOptionLevel', y='Count', hue='Attrition',data = stock_att, palette='ch:s=.25,rot=-.25')

# Calculate and add Attrition rates
stock_count = pd.DataFrame(pd.crosstab(df['StockOptionLevel'],df['Attrition']))
stock_count.insert(2, 'Attr%', value = (stockoption_yes/(stockoption_no + stockoption_yes) *100))
stock_count.insert(2, 'Count', value = stock_option_count)

# Specify that we want to share the same x-axis
ax2 = ax1.twinx()
color = 'tab:orange'

# Line plot creation
ax2.set_ylabel('% Attrition', fontsize=15, color=color)
ax2 = sns.lineplot(x='StockOptionLevel', y='Attr%', data=stock_count, sort=False, color=color)
ax2.tick_params(axis='y', color=color)

plt.show()
```

```python
##Observations: 
1. Attrition is affected by Monthly Income and Stock Option Levels.
2. The highest attrition was demonstrated by those earning less monthly income and lower job levels.
3. Sales Representatives have the highest attrition rate followed by Laboratory Technicians. 
```
